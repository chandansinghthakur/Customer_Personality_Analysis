# Data cleaning and wrangling

```python
# Removing the null values from dataset
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
# Dt_Customer has date in the dataset but it is not parsed correct dtype as Datetime

df['Dt_Customer']=pd.to_datetime(df['Dt_Customer'])
```

```python
df.info()
```

```python
df['Dt_Customer']=df['Dt_Customer'].astype('str')
```

```python
nowdate = '2023-01-11'
df['Customer_days_enrollment']=0
for i in df.index:
    df['Dt_Customer'][i]
    d1 = datetime.strptime(nowdate, "%Y-%m-%d")
    d2 = datetime.strptime(df['Dt_Customer'][i], "%Y-%m-%d")
    delta = d1-d2
    df['Customer_days_enrollment'][i]=delta.days

df['Customer_days_enrollment']=df['Customer_days_enrollment'].astype('int')
```

```python
# extracting age of customer from Year_Birth feature

df['Age'] = 2023- df['Year_Birth']
```

```python
# creating spent feature which indicates total spending of a custmer on items

df['Spent']=df['MntWines'] + df['MntFruits'] +df['MntMeatProducts'] +df['MntFishProducts'] +df['MntSweetProducts'] +df['MntGoldProds'] 
```

```python
# creating a new feature which indicates of a customer is in relationship or not

df['Marital_Status'].unique()
```

```python
df['In_relationship'] = df['Marital_Status'].replace({
    'Single':0, 'Together':1, 'Married':1, 'Divorced':0, 'Widow':0, 'Alone':0,
    'Absurd':0, 'YOLO':0
})

df['In_relationship']=df['In_relationship'].astype('int')
```

```python
df.head()
```

```python
# creating feature indicating total number of children living with customer

df['Children'] = df['Kidhome'] + df['Teenhome']
```

```python
# Customer family size

df['Family_size']=df['In_relationship']+ df['Children'] +1
```

```python
# updating education feature

df['Education'].unique()
```

```python
df['Education']=df['Education'].replace({
    'Graduation':'Graduate', 'PhD':'Postgraduate', 'Master':'Postgraduate', 'Basic': 'Undergraduate', '2n Cycle': 'Undergraduate'
})
```

```python
# drop extra features

drop_columns = ['Marital_Status', 'Income', 'Kidhome',
       'Teenhome', 'Dt_Customer','Z_CostContact', 'Z_Revenue','ID']

df.drop(drop_columns, axis=1, inplace=True)
```
